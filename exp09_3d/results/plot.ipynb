```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('./vgg16.pkl', 'rb') as f:
    vgg16_scores = pickle.load(f)

with open('./vit_base.pkl', 'rb') as f:
    vit_base_scores = pickle.load(f)
    
```

```python
vgg16_scores.shape
```

```python
ref_1_score = 0.4
ref_1_name = 'Enns & Rensink,1990'

ref_2_score = 0.76
ref_2_name = 'Enns & Rensink,1991'
```

```python
sns.set_style('whitegrid')
colors = sns.color_palette()
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 5))

ax1_layer_inds = np.arange(len(vgg16_scores[:,0]))
sns.lineplot(x=ax1_layer_inds, y=vgg16_scores[:,0], ax=ax1)
ax1.axhline(y=ref_1_score, color=colors[0], linestyle='--')
ax1.text(0, ref_1_score+0.1, ref_1_name, verticalalignment='top', horizontalalignment='left')
sns.lineplot(x=ax1_layer_inds, y=vgg16_scores[:,1], ax=ax1)
ax1.axhline(y=ref_2_score, color=colors[1], linestyle='--')
ax1.text(0, ref_2_score+0.1, ref_2_name, verticalalignment='top', horizontalalignment='left')

ax1.set_ylim(-1,1)
ax1.set_title('VGG16')


ax2_layer_inds = np.arange(len(vit_base_scores[:,0]))
sns.lineplot(x=ax2_layer_inds, y=vit_base_scores[:,0], ax=ax2)
ax2.axhline(y=ref_1_score, color=colors[0], linestyle='--')
ax2.text(0, ref_1_score+0.1, ref_1_name, verticalalignment='top', horizontalalignment='left')
sns.lineplot(x=ax2_layer_inds, y=vit_base_scores[:,1], ax=ax2)
ax2.axhline(y=ref_2_score, color=colors[1], linestyle='--')
ax2.text(0, ref_2_score+0.1, ref_2_name, verticalalignment='top', horizontalalignment='left')

ax2.set_ylim(-1,1)
ax2.set_title('ViT')



fig.supxlabel('Layers')
fig.supylabel("3D Processing Index")



plt.show()
```
