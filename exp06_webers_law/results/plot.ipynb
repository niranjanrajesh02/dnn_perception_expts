```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('./vgg16_correlations.pkl', 'rb') as f:
    vgg16_correlations = pickle.load(f)

with open('./vit_base_correlations.pkl', 'rb') as f:
    vit_correlations = pickle.load(f)
```

```python
reference_mi_Length=0.1;
reference_name ='Pramod and Arun, 2014';
```

```python
sns.set_style('whitegrid')
colors = sns.color_palette()
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 5))

ax1_layer_inds = np.arange(len(vgg16_correlations))
sns.lineplot(x=ax1_layer_inds, y=vgg16_correlations, ax=ax1)
ax1.axhline(y=reference_mi_Length, color=colors[0], linestyle='--')
ax1.text(0, reference_mi_Length+0.1, reference_name, verticalalignment='top', horizontalalignment='left')


ax1.set_ylim(-1,1)
ax1.set_title('VGG16')

ax2_layer_inds = np.arange(len(vit_correlations))
sns.lineplot(x=ax2_layer_inds, y=vit_correlations, ax=ax2)
ax2.axhline(y=reference_mi_Length, color=colors[0], linestyle='--')
ax2.text(0, reference_mi_Length+0.1, reference_name, verticalalignment='top', horizontalalignment='left')

ax2.set_ylim(-1,1)
ax2.set_title('ViT Base')


fig.supxlabel('Layers')
fig.supylabel("Relative length correlation - Absolute length correlation")



plt.show()
```
