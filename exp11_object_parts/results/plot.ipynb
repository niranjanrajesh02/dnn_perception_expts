```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('./vgg16_part_matching.pkl'):
    vgg16_part_matching = pickle.load(open('./vgg16_part_matching.pkl', 'rb'))

with open('./vit_base_part_matching.pkl'):
    vit_base_part_matching = pickle.load(open('./vit_base_part_matching.pkl', 'rb'))
```

```python
reference_pm= 0.071;
reference_name='Xu and Singh, 2002';
```

```python
sns.set_style('whitegrid')
colors = sns.color_palette()
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 5))

ax1_layer_inds = np.arange(len(vgg16_part_matching))
sns.lineplot(x=ax1_layer_inds, y=vgg16_part_matching, ax=ax1)
ax1.axhline(y=reference_pm, color=colors[0], linestyle='--')
ax1.text(0, reference_pm+0.1, reference_name, verticalalignment='top', horizontalalignment='left')


ax1.set_ylim(-1,1)
ax1.set_title('VGG16')

ax2_layer_inds = np.arange(len(vit_base_part_matching))
sns.lineplot(x=ax2_layer_inds, y=vit_base_part_matching, ax=ax2)

ax2.axhline(y=reference_pm, color=colors[0], linestyle='--')
ax2.text(0, reference_pm+0.1, reference_name, verticalalignment='top', horizontalalignment='left')


ax2.set_ylim(-1,1)
ax2.set_title('ViT-Base')

fig.supxlabel('Layers')
fig.supylabel("Part-Matching Index")



plt.show()
```
