```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pickle
```

```python
with open('./morph_ref_vgg16.pkl', 'rb') as f:
    morph_ref_vgg16 = pickle.load(f)

with open('./morph_ref_vit_base.pkl', 'rb') as f:
     morph_ref_vit_base = pickle.load(f)

with open('./shape_texture_vgg16.pkl', 'rb') as f:
    shape_texture_vgg16 = pickle.load(f)

with open('./shape_texture_vit_base.pkl', 'rb') as f:
     shape_texture_vit_base = pickle.load(f)
```

```python
reference_mi_morph=0.73; 
reference_mi_shapetexture=0.68; 
reference_name='Zhivago and Arun, 2016';
```

```python
shape_texture_vit_base
```

```python
np.arange(len(morph_ref_vgg16))
```

```python
sns.set_style('whitegrid')
colors = sns.color_palette()
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 5))

ax1_layer_inds = np.arange(len(morph_ref_vgg16))
sns.lineplot(x=ax1_layer_inds, y=morph_ref_vgg16, ax=ax1)
ax1.axhline(y=reference_mi_morph, color=colors[0], linestyle='--')
ax1.text(0, reference_mi_morph+0.2, reference_name, verticalalignment='top', horizontalalignment='left', fontsize=12)


ax1.set_ylim(-1,1)
ax1.set_title('VGG16')



ax2_layer_inds = np.arange(len(morph_ref_vit_base))
sns.lineplot(x=ax2_layer_inds, y=morph_ref_vit_base, ax=ax2)
ax2.axhline(y=reference_mi_morph, color=colors[0], linestyle='--')
ax2.text(0, reference_mi_morph+0.1, reference_name, verticalalignment='top', horizontalalignment='left')

ax2.set_ylim(-1,1)
ax2.set_title('ViT Base')


fig.supxlabel('Layers')
fig.supylabel("Sparseness correlation across sets")
fig.suptitle("Reference set vs Morphlines")


plt.show()
```

```python
sns.set_style('whitegrid')
colors = sns.color_palette()
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 5))

ax1_layer_inds = np.arange(len(shape_texture_vgg16))
sns.lineplot(x=ax1_layer_inds, y=shape_texture_vgg16, ax=ax1)
ax1.axhline(y=reference_mi_shapetexture, color=colors[0], linestyle='--')
ax1.text(0, reference_mi_shapetexture+0.2, reference_name, verticalalignment='top', horizontalalignment='left', fontsize=12)


ax1.set_ylim(-1,1)
ax1.set_title('VGG16')

ax2_layer_inds = np.arange(len(shape_texture_vit_base))
sns.lineplot(x=ax2_layer_inds, y=shape_texture_vit_base, ax=ax2)
ax2.axhline(y=reference_mi_shapetexture, color=colors[0], linestyle='--')
ax2.text(0, reference_mi_shapetexture+0.1, reference_name, verticalalignment='top', horizontalalignment='left')

ax2.set_ylim(-1,1)
ax2.set_title('ViT Base')


fig.supxlabel('Layers')
fig.supylabel("Sparseness correlation across sets")
fig.suptitle("Shapes vs Textures")


plt.show()
```
