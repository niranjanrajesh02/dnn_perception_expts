```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('./vgg16_si.pkl', 'rb') as f:
    vgg16_scores = pickle.load(f)

with open('./vit_base_si.pkl', 'rb') as f:
    vit_base_scores = pickle.load(f)
    
```

```python
reference_mi=0.55;
reference_name='Ratan Murty & Arun 2016';
```

```python
sns.set_style('whitegrid')
colors = sns.color_palette()
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 5))

ax1_layer_inds = np.arange(len(vgg16_scores))
sns.lineplot(x=ax1_layer_inds, y=vgg16_scores, ax=ax1)
ax1.axhline(y=reference_mi, color=colors[0], linestyle='--')
ax1.text(0, reference_mi+0.1, reference_name, verticalalignment='top', horizontalalignment='left')

ax1.set_ylim(-1,1)
ax1.set_title('VGG16')


ax2_layer_inds = np.arange(len(vit_base_scores))
sns.lineplot(x=ax2_layer_inds, y=vit_base_scores, ax=ax2)
ax2.axhline(y=reference_mi, color=colors[0], linestyle='--')
ax2.text(0, reference_mi+0.1, reference_name, verticalalignment='top', horizontalalignment='left')


ax2.set_ylim(-1,1)
ax2.set_title('ViT')



fig.supxlabel('Layers')
fig.supylabel("Average Surface Invariance Index for top 7 % tetrad units")


plt.show()
```
