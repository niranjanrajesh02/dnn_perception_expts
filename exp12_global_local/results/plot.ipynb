```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('./vgg16_global_adv.pkl'):
    vgg16_global_adv = pickle.load(open('./vgg16_global_adv.pkl', 'rb'))

with open('./vit_base_global_adv.pkl'):
    vit_base_global_adv = pickle.load(open('./vit_base_global_adv.pkl', 'rb'))
```

```python
reference_global_adv=0.0975;
reference_name='Jacob and Arun, 2019';
```

```python
sns.set_style('whitegrid')
colors = sns.color_palette()
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 5))

ax1_layer_inds = np.arange(len(vgg16_global_adv))
sns.lineplot(x=ax1_layer_inds, y=vgg16_global_adv, ax=ax1)
ax1.axhline(y=reference_global_adv, color=colors[0], linestyle='--')
ax1.text(0, reference_global_adv+0.1, reference_name, verticalalignment='top', horizontalalignment='left')


ax1.set_ylim(-1,1)
ax1.set_title('VGG16')

ax2_layer_inds = np.arange(len(vit_base_global_adv))
sns.lineplot(x=ax2_layer_inds, y=vit_base_global_adv, ax=ax2)

ax2.axhline(y=reference_global_adv, color=colors[0], linestyle='--')
ax2.text(0, reference_global_adv+0.1, reference_name, verticalalignment='top', horizontalalignment='left')


ax2.set_ylim(-1,1)
ax2.set_title('ViT-Base')

fig.supxlabel('Layers')
fig.supylabel("Global Advantage Index")



plt.show()
```
